```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Make sure we have the correct directory
print(os.getcwd())
```

```python
#Load the popular_tracks.csv and audio_features datasets into dataframes
df_audio_features = pd.read_csv("audio_features.csv")
df_popular_tracks = pd.read_csv("popular_tracks.csv")
```

```python
df_popular_tracks.head()
```

```python
df_audio_features.head()
```

```python
df_popular_tracks.info()
```

```python
df_audio_features.info()
```

```python
df_popular_tracks_w_audio_features = pd.merge(df_popular_tracks, df_audio_features, on="isrc", how='inner' )
```

```python
df_popular_tracks_w_audio_features.info()
```

```python
df_popular_tracks_w_audio_features.head()
```

```python
# Load sp_artist and sp_artist_release and merge everthing together

df_sp_artist = pd.read_csv("sp_artist.csv")
df_sp_artist_release = pd.read_csv("sp_artist_release.csv")
```

```python
df_sp_artist.info()
```

```python
df_sp_artist_release.info()
```

```python
df_sp_artist = pd.merge(df_sp_artist, df_sp_artist_release, on='artist_id', how='left')
df_sp_artist.info()
```

```python
# Now merge everything to produce the final set
df_final_set = pd.merge(df_popular_tracks_w_audio_features, df_sp_artist, on='release_id', how='left')
df_final_set.info()
```

```python
df_final_set.head(50)
```

```python
#Remove duplicating rows
mask = df_final_set.duplicated(subset='isrc', keep='first')
df_final_set = df_final_set[~mask]
df_final_set.info()
```

```python
print(os.getcwd())
```

```python
df_final_set.to_csv('final_set_w-out_popularity.csv')
```
